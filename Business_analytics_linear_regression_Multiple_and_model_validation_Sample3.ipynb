1.1. LINEAR REGRESSION - CODE SNIPPETS


1.2.   Marketing dataset: Advertising Spend (in thousands) and Sales
1.3.   Run Multiple Linear Regression




```r
# Marketing dataset: Advertising Spend (in thousands) and Sales
marketing_data <- data.frame(
  TV = c(230, 44, 17, 151, 180, 8, 199, 66, 214, 23),
  Radio = c(37, 39, 45, 41, 10, 48, 3, 9, 14, 35),
  Newspaper = c(69, 45, 69, 58, 14, 23, 34, 48, 27, 12),
  Sales = c(22.1, 10.4, 9.3, 18.5, 12.9, 7.2, 12.8, 8.7, 17.1, 9.5)
)
```

```r
head(marketing_data)
```

1.3. Run Multiple Linear Regression

```r
tail(marketing_data)
```

```r
summary.data.frame(marketing_data)
```

```r
# Linear regression model
model <- lm(Sales ~ TV + Radio + Newspaper, data = marketing_data)

# View model summary
summary(model)
```

```r
# New marketing plan
new_spend <- data.frame(
  TV = 200,
  Radio = 30,
  Newspaper = 20
)

# Predict sales
predict(model, new_spend)
```

```r
plot(marketing_data$TV, marketing_data$Sales,
     main = "TV Advertising vs Sales",
     xlab = "TV Spend",
     ylab = "Sales")

abline(lm(Sales ~ TV, data = marketing_data), col = "blue", lwd = 2)
```
