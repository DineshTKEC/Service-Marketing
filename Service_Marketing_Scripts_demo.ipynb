**Synthetic dataset for a retail outlet with ratings on various aspects of the 7Ps of the Service Marketing Mix. For this example, we will use a dataset with the following columns:**

 *Store *italicized text* ID: The identifier for different retail outlets.

    Product Quality: Rating of product quality (1-10 scale).

    Price: Rating of the perceived value for the price (1-10 scale).

    Store Location: Rating of the store’s location (1-10 scale).

    Promotion Effectiveness: Rating of the effectiveness of promotions (1-10 scale).

    Employee Service: Rating of employee service (1-10 scale).

    Process Efficiency: Rating of the efficiency of store processes (1-10 scale).

    Store Environment: Rating of the store’s physical environment (1-10 scale). *italicized text** *italicized text*

```r
# Load necessary libraries
library(dplyr)
library(ggplot2)

# Create a hypothetical retail outlet dataset
set.seed(123)  # For reproducibility

# Simulating data for 50 retail outlets
retail_data <- data.frame(
  Store_ID = 1:50,
  Product_Quality = sample(1:10, 50, replace = TRUE),
  Price = sample(1:10, 50, replace = TRUE),
  Store_Location = sample(1:10, 50, replace = TRUE),
  Promotion_Effectiveness = sample(1:10, 50, replace = TRUE),
  Employee_Service = sample(1:10, 50, replace = TRUE),
  Process_Efficiency = sample(1:10, 50, replace = TRUE),
  Store_Environment = sample(1:10, 50, replace = TRUE)
)

# View the first few rows of the dataset
head(retail_data)
```

```r
# Load necessary libraries
library(dplyr)
library(tidyr)  # Make sure tidyr is loaded
library(ggplot2)

# Create a hypothetical retail outlet dataset
set.seed(123)  # For reproducibility

# Simulating data for 50 retail outlets
retail_data <- data.frame(
  Store_ID = 1:50,
  Product_Quality = sample(1:10, 50, replace = TRUE),
  Price = sample(1:10, 50, replace = TRUE),
  Store_Location = sample(1:10, 50, replace = TRUE),
  Promotion_Effectiveness = sample(1:10, 50, replace = TRUE),
  Employee_Service = sample(1:10, 50, replace = TRUE),
  Process_Efficiency = sample(1:10, 50, replace = TRUE),
  Store_Environment = sample(1:10, 50, replace = TRUE)
)

# View the first few rows of the dataset
head(retail_data)

# Reshape the data using pivot_longer (replaces gather())
retail_data_long <- retail_data %>%
  pivot_longer(cols = -Store_ID, names_to = "Aspect", values_to = "Rating")

# Plot the ratings across the 7Ps
ggplot(retail_data_long, aes(x = Aspect, y = Rating, fill = Aspect)) +
  geom_boxplot() +
  theme_minimal() +
  labs(title = "Evaluation of Service Marketing Mix for Retail Outlets",
       x = "Service Marketing Mix Aspect",
       y = "Rating (1-10)") +
  theme(axis.text.x = element_text(angle = 45, hjust = 1))
```

```r
# Calculate the average rating for each aspect
average_ratings <- retail_data %>%
  summarise(
    Product_Quality = mean(Product_Quality),
    Price = mean(Price),
    Store_Location = mean(Store_Location),
    Promotion_Effectiveness = mean(Promotion_Effectiveness),
    Employee_Service = mean(Employee_Service),
    Process_Efficiency = mean(Process_Efficiency),
    Store_Environment = mean(Store_Environment)
  )

# Print the average ratings
average_ratings
```
