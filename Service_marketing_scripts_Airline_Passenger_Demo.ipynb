```r
# Step 1: Load and Explore the Dataset
data("AirPassengers")
summary(AirPassengers)
class(AirPassengers)

# Step 2: Visualize the Time Series
plot(AirPassengers,
     main = "Monthly Air Passengers (1949-1960)",
     ylab = "Number of Passengers",
     xlab = "Year",
     col = "blue")

# Step 3: Decompose the Time Series
decomposed <- decompose(AirPassengers)
plot(decomposed)

# Step 4: Forecasting with Holt-Winters Exponential Smoothing
hw_model <- HoltWinters(AirPassengers)
plot(hw_model)

# Forecasting the next 12 months
library(forecast)
forecast_hw <- forecast(hw_model, h = 12)
plot(forecast_hw)

# Step 5: Interpretation
# Check seasonal strength and trend growth
seasonplot(AirPassengers, year.labels=TRUE, col=rainbow(12), main="Seasonal Plot")
```

```r
# Load required libraries
library(forecast)
library(ggplot2)

# Load dataset
data("AirPassengers")

# Plot original time series
plot(AirPassengers,
     main = "Monthly Air Passengers (1949–1960)",
     ylab = "Number of Passengers",
     xlab = "Year",
     col = "blue")

# Decompose the time series
decomp <- decompose(AirPassengers)
plot(decomp)

# Holt-Winters model and forecast
hw_model <- HoltWinters(AirPassengers)
plot(hw_model)

forecast_values <- forecast(hw_model, h = 12)
plot(forecast_values)

# Seasonal plot
seasonplot(AirPassengers,
           year.labels = TRUE,
           col = rainbow(12),
           main = "Seasonal Pattern of Air Travel")

```
